```
from bs4 import BeautifulSoup
import requests
data = []
url = 'https://www.imdb.com/search/title/?title_type=feature,tv_movie&countries=ir&languages=fa&sort=alpha,asc&ref_=adv_prv'
resp = requests.get(url)
soup = BeautifulSoup(resp.text, "lxml")
# print(soup)
llist = soup.find_all('h3',class_='lister-item-header')
print(len(llist))
data2 = []
for x in llist:
     for y in x.find_all('a', href=True):
          data2.append(str(y['href'][9:-1]))
          data.append(y.text)
          print(y.text)
for i in range(1,75):
    url = 'https://www.imdb.com/search/title/?title_type=feature,tv_movie&countries=ir&languages=fa&sort=alpha,asc&start='+str(i*50 + 1)+'&ref_=adv_nxt'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, "lxml")
    # print(soup)
    llist = soup.find_all('h3',class_='lister-item-header')
    for x in llist:
         for y in x.find_all('a'):
             data2.append(str(y['href'][9:-1]))
             data.append(y.text)
             print(y.text)
```

```
with open("data1.txt", "w") as file:

    file_lines = "\n".join(data2)
    file.write(file_lines)
file.close()
```

```
# !pip install imdbpy
import imdb
import random
# from genderize import Genderize
ia = imdb.IMDb()

my_file = open("data1.txt", "r")
content_list = my_file.readlines()
# print(content_list)
lis = []
print(len(content_list))
for i in content_list:
    x = i.rstrip()
    lis.append(x)
moviesListActor = []
actors = []
allActors = []
years = [i for i in range(1940, 1998)]
for i in lis:
    # movies = ia.search_movie(lis[i])
    # id = movies[0].movieID
    movie = ia.get_movie(i)
    cast = None
    year = None
    rating = None
    genre = None
    try:
      cast = movie['cast']
      rating = movie.data['rating']
      year = movie.data['year']  
      genre = movie['genre'][0]
    except:
      continue 
    for j in range(len(cast)):
        yob = None
        try:
          yob = ia.get_person(cast[j].personID)['birth date']
        except:
          yob = None
        lists = []
        # # gender = Genderize().get([cast[j]['name'].split(" ")[0]])[0]['gender']
        # if gender == None:
        #   gender = random.choice(['male', 'female'])
        nm = cast[j]['name']
        allActors.append(nm)
        l = [item for item in actors if item[0] == nm]
        if yob == None:  
          if len(l) == 0:
            yob = str(random.choice(years))
            lists = [i, movie['title'], genre,year,rating, cast[j].personID, nm.split(" ")[0], " ".join(nm.split(" ")[1:]),  yob,str(cast[j].currentRole)]
            actors.append((nm, str(yob)))
          else:
            yob = l[0][1]
            lists = [i, movie['title'], genre, year, rating, cast[j].personID, nm.split(" ")[0], " ".join(nm.split(" ")[1:]), yob  ,str(cast[j].currentRole)]
        else:
          lists = [i, movie['title'], genre, year, rating, cast[j].personID, nm.split(" ")[0], " ".join(nm.split(" ")[1:]), yob[0:4],str(cast[j].currentRole)]
        moviesListActor.append(lists)
        print(lists)
```

```
from faker import Faker
import random

fake = Faker()

doctors_count = 400
patient_count = 800
medicines_count = 3051
prescriptions_count = 4000
prescription_medicine_count = 4000
phone_numbers = random.sample(range(10 ** 8, 10 ** 9), doctors_count)
ssns = random.sample(range(10 ** 7, 10 ** 8), doctors_count + patient_count)
yearOfExprience = [random.randint(1, 40) for _ in range(doctors_count)]
prescriptionsId = random.sample(range(10**4, 10**5), prescriptions_count)

prescription_medicine = []
prescriptions = []
medicines = []
doctors = [[ssns[0], 'ali', 'irani', random.choice(special), yearOfExprience[0], phone_numbers[0]]]
patients = []
for i in range(1,doctors_count):
  temp = [ssns[i], fake.first_name(), fake.last_name(), random.choice(special), yearOfExprience[i], phone_numbers[i]]
  doctors.append(temp)

for j in range(patient_count):
  temp = [ssns[j+doctors_count], fake.first_name(), fake.last_name(), fake.address().replace('\n', '-'), fake.date(), ssns[j%doctors_count]]
  patients.append(temp)
for k in range(medicines_count):
  temp = [medici[k], random.randint(4, 40), random.choice([True, False])]
  medicines.append(temp)

for l in range(prescriptions_count):
  tmp = random.choice(patients)
  temp = [prescriptionsId[l], fake.date(), tmp[-1], tmp[0]]
  prescriptions.append(temp)

for m in range(prescriptions_count):
    ID = prescriptionsId[m]
    for n in range(random.randint(1, 4)):
      temp = [ID, random.choice(medici), random.randint(5, 120)]
      prescription_medicine.append(temp)





```
